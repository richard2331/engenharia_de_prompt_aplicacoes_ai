```python
import random # Importa o módulo 'random' para gerar números aleatórios

# Simula a leitura da temperatura do sistema
# Em um cenário real, esta variável seria lida de um sensor físico.
# Geramos um valor aleatório entre 20.0 e 40.0 para simular.
temperature = random.uniform(20.0, 40.0)

# Exibe a temperatura atual do sistema utilizando concatenação de strings
# O método '%.2f' formata o número de ponto flutuante para duas casas decimais.
print("Temperatura atual do sistema: " + "%.2f" % temperature + "°C")

# Verifica se a temperatura ultrapassa o limite de segurança (30°C)
if temperature > 30:
    # Se a temperatura for maior que 30°C, exibe um alerta de superaquecimento
    print("ALERTA: Superaquecimento detectado! Temperatura acima de 30°C.")
else:
    # Caso contrário, informa que a temperatura está dentro dos limites normais
    print("Temperatura do sistema está normal.")
```

```python
import os     # Importa o módulo 'os' para interagir com o sistema operacional (caminhos, diretórios, etc.)
import shutil   # Importa o módulo 'shutil' para operações de alto nível em arquivos e diretórios (mover, copiar, apagar)
import random   # Importa o módulo 'random' para gerar nomes aleatórios para os arquivos de teste

def organize_files(source_folder):
    # Define um dicionário que mapeia extensões de arquivo a nomes de pastas de destino.
    # Isso permite que o programa saiba para onde mover cada tipo de arquivo.
    file_types = {
        '.pdf': 'PDF',
        '.jpg': 'Imagens', '.jpeg': 'Imagens', '.png': 'Imagens', '.gif': 'Imagens',
        '.xls': 'Planilhas', '.xlsx': 'Planilhas', '.csv': 'Planilhas',
        '.doc': 'Documentos', '.docx': 'Documentos', '.txt': 'Documentos',
        '.py': 'Codigo', '.ipynb': 'Jupyter Notebooks', '.java': 'Codigo', '.cpp': 'Codigo',
        '.zip': 'Compactados', '.rar': 'Compactados', '.7z': 'Compactados',
        '.mp3': 'Audio', '.wav': 'Audio', '.aac': 'Audio',
        '.mp4': 'Videos', '.mov': 'Videos', '.avi': 'Videos',
        # Você pode adicionar mais extensões e categorias conforme sua necessidade.
    }

    # ---------------------------------------------------
    # Execução: Verificação da pasta de origem
    # ---------------------------------------------------
    # Verifica se o caminho da pasta de origem fornecido pelo usuário realmente existe.
    # Se não existir, imprime uma mensagem de erro e interrompe a execução da função.
    if not os.path.isdir(source_folder):
        print(f"Erro: A pasta de origem '{source_folder}' não existe. Por favor, verifique o caminho fornecido.")
        return

    print(f"\n--- Iniciando a organização dos arquivos na pasta: '{source_folder}' ---")

    # ---------------------------------------------------
    # Execução: Listagem e processamento de itens na pasta
    # ---------------------------------------------------
    # Itera sobre cada item (arquivo ou pasta) encontrado dentro da pasta de origem.
    for item in os.listdir(source_folder):
        item_path = os.path.join(source_folder, item) # Constrói o caminho completo para o item atual

        # ---------------------------------------------------
        # Execução: Ignorando diretórios e o próprio script
        # ---------------------------------------------------
        # Se o item for um diretório (uma pasta), ele é ignorado para evitar que o programa tente movê-lo
        # como se fosse um arquivo ou processe seu conteúdo de forma inadequada.
        # Também ignora o próprio script (se por acaso estivesse dentro da pasta a ser organizada).
        if os.path.isdir(item_path):
            print(f"  [Ignorado]: '{item}' é um diretório existente.")
            continue

        # ---------------------------------------------------
        # Execução: Extração da extensão do arquivo
        # ---------------------------------------------------
        # Usa os.path.splitext para separar o nome do arquivo da sua extensão.
        # Ex: "documento.pdf" -> ("documento", ".pdf")
        file_name, file_extension = os.path.splitext(item)
        file_extension = file_extension.lower() # Converte a extensão para minúsculas para garantir consistência (ex: .PDF é tratado como .pdf)

        # ---------------------------------------------------
        # Execução: Determinação da pasta de destino
        # ---------------------------------------------------
        # Tenta encontrar a categoria da pasta correspondente à extensão do arquivo no dicionário `file_types`.
        # Se a extensão não for encontrada (ou seja, não está mapeada), o arquivo será movido para a pasta 'Outros'.
        target_folder_name = file_types.get(file_extension, 'Outros')
        target_path = os.path.join(source_folder, target_folder_name) # Constrói o caminho completo da pasta de destino

        # ---------------------------------------------------
        # Execução: Criação da pasta de destino (se necessário)
        # ---------------------------------------------------
        # Verifica se a pasta de destino para a categoria atual já existe.
        # Se não existir, os.makedirs a cria. `exist_ok=True` impede que o programa gere um erro caso a pasta já exista.
        if not os.path.exists(target_path):
            os.makedirs(target_path)
            print(f"  [Criada]: Pasta '{target_folder_name}' em '{source_folder}'.")

        # ---------------------------------------------------
        # Execução: Movendo o arquivo
        # ---------------------------------------------------
        # Constrói o caminho completo para onde o arquivo será movido (incluindo o nome do arquivo).
        destination_path = os.path.join(target_path, item)

        try:
            # Tenta mover o arquivo do seu local original para a pasta de destino.
            shutil.move(item_path, destination_path)
            print(f"  [Movido]: Arquivo '{item}' para '{target_folder_name}'.")
        except shutil.Error as e:
            # Captura erros específicos de shutil (ex: permissão negada, arquivo já existe no destino).
            print(f"  [ERRO ao mover]: Arquivo '{item}' para '{target_folder_name}': {e}")
        except Exception as e:
            # Captura quaisquer outros erros inesperados durante a operação de mover.
            print(f"  [ERRO inesperado]: Ao mover '{item}': {e}")

    print("\n--- Organização de arquivos concluída com sucesso! ---")


# --- Seção para Demonstração com Geração de Arquivos de Teste ---
# Esta parte do código cria uma pasta temporária, gera arquivos falsos e os organiza.

# Define o nome da pasta de teste que será criada e usada para a demonstração.
TEST_FOLDER_NAME = "pasta_teste_organizacao"
# Constrói o caminho completo para a pasta de teste no diretório de trabalho atual.
TEST_FOLDER_PATH = os.path.join(os.getcwd(), TEST_FOLDER_NAME)

# Define os tipos de arquivos e a quantidade a serem gerados para o teste.
TEST_FILE_TYPES = {
    'documento': '.doc',
    'imagem': '.jpg',
    'planilha': '.xls',
    'pdf': '.pdf',
    'codigo': '.py',
    'notebook': '.ipynb'
}
NUM_FILES_PER_TYPE = 10 # Quantidade de arquivos a serem gerados para cada tipo

# ---------------------------------------------------
# Execução: Preparação do ambiente de teste
# ---------------------------------------------------
print(f"\nPreparando o ambiente de teste em: {TEST_FOLDER_PATH}")

# Limpa a pasta de teste se ela já existir para garantir um teste 'limpo'.
if os.path.exists(TEST_FOLDER_PATH):
    print(f"  [Limpeza]: Removendo pasta de teste existente: '{TEST_FOLDER_NAME}'.")
    shutil.rmtree(TEST_FOLDER_PATH)

# Cria a pasta de teste.
os.makedirs(TEST_FOLDER_PATH)
print(f"  [Criada]: Pasta de teste: '{TEST_FOLDER_NAME}'.")

# ---------------------------------------------------
# Execução: Geração de arquivos de teste
# ---------------------------------------------------
print(f"  [Gerando]: {NUM_FILES_PER_TYPE} arquivos para cada tipo de teste...")
for file_base_name, ext in TEST_FILE_TYPES.items():
    for i in range(NUM_FILES_PER_TYPE):
        # Gera um nome de arquivo único para cada tipo e número.
        file_name = f"{file_base_name}_{i+1}{ext}"
        file_path = os.path.join(TEST_FOLDER_PATH, file_name)
        # Cria um arquivo vazio (simula um arquivo real).
        with open(file_path, 'w') as f:
            f.write(f"Conteúdo de teste para {file_name}")
        # print(f"    Criado: {file_name}") # Descomente para ver cada arquivo sendo criado
print(f"  [Gerados]: {len(TEST_FILE_TYPES) * NUM_FILES_PER_TYPE} arquivos de teste com sucesso.")

# ---------------------------------------------------
# Execução: Chamada da função de organização
# ---------------------------------------------------
# Chama a função principal `organize_files` para organizar os arquivos na pasta de teste.
organize_files(TEST_FOLDER_PATH)

# ---------------------------------------------------
# Execução: Verificação dos resultados da organização
# ---------------------------------------------------
print(f"\n--- Verificando a organização na pasta: '{TEST_FOLDER_NAME}' ---")
for root, dirs, files in os.walk(TEST_FOLDER_PATH):
    level = root.replace(TEST_FOLDER_PATH, '').count(os.sep)
    indent = ' ' * 4 * (level)
    print(f'{indent}{os.path.basename(root)}/')
    subindent = ' ' * 4 * (level + 1)
    for f in files:
        print(f'{subindent}{f}')

# ---------------------------------------------------
# Execução: Limpeza do ambiente de teste
# ---------------------------------------------------
# Remove a pasta de teste e todos os seus conteúdos após a demonstração.
# Isso garante que nenhum arquivo temporário seja deixado no sistema.
print(f"\n--- Limpeza: Removendo pasta de teste: '{TEST_FOLDER_NAME}' ---")
if os.path.exists(TEST_FOLDER_PATH):
    shutil.rmtree(TEST_FOLDER_PATH)
    print(f"  [Removida]: Pasta de teste '{TEST_FOLDER_NAME}' e seus conteúdos foram apagados.")
else:
    print(f"  [Alerta]: Pasta de teste '{TEST_FOLDER_NAME}' não encontrada para limpeza.")

print("Demonstração de organização de arquivos concluída e ambiente limpo.")
```

```python
import requests # Importa a biblioteca 'requests' para fazer requisições HTTP (para a API do CEP)

def get_address_by_cep(cep):
    # Remove quaisquer caracteres não numéricos do CEP para garantir o formato correto
    cep = cep.replace("-", "").replace(".", "").replace(" ", "")

    # Verifica se o CEP tem 8 dígitos, que é o formato padrão brasileiro
    if len(cep) != 8 or not cep.isdigit():
        return {"erro": True, "mensagem": "CEP inválido. Deve conter 8 dígitos numéricos."}

    # URL da BrasilAPI v2, que agrega dados de diversas fontes, incluindo os Correios.
    # Uma API pública direta dos Correios para consulta geral de CEP não está amplamente disponível.
    url = f"https://brasilapi.com.br/api/cep/v2/{cep}"

    try:
        # Faz a requisição GET para a API da BrasilAPI
        response = requests.get(url)

        # Lança um erro HTTP para status de erro (4xx ou 5xx).
        # Isso cobre casos como CEP não encontrado (404) ou outros problemas do servidor.
        response.raise_for_status()

        # Converte a resposta JSON da API em um dicionário Python
        data = response.json()

        # A BrasilAPI, mesmo com status 200, pode retornar um objeto de erro se o CEP não for válido internamente.
        # Se `data` é um dicionário e contém 'message' e 'name', geralmente indica um erro da API.
        if isinstance(data, dict) and 'message' in data and 'name' in data:
            return {"erro": True, "mensagem": f"Erro na consulta do CEP: {data['message']} ({data['name']})"}
        else:
            return {"erro": False, "data": data}

    except requests.exceptions.HTTPError as e:
        # Captura erros HTTP específicos (e.g., 404 Not Found, 500 Internal Server Error)
        if e.response.status_code == 404:
            return {"erro": True, "mensagem": f"CEP '{cep}' não encontrado."}
        else:
            return {"erro": True, "mensagem": f"Erro na requisição HTTP: {e}"}
    except requests.exceptions.RequestException as e:
        # Captura outros erros relacionados à requisição (ex: problemas de conexão, DNS)
        return {"erro": True, "mensagem": f"Erro de conexão: {e}"}
    except ValueError:
        # Captura erros se a resposta não for um JSON válido
        return {"erro": True, "mensagem": "Resposta inválida da API (não é JSON)."}

# --- Bloco principal de execução ---

# Solicita ao usuário que digite um CEP para consulta
cep_input = input("Digite o CEP (apenas números ou com hífen): ")

# Chama a função para obter as informações do endereço
result = get_address_by_cep(cep_input)

# Verifica o resultado retornado pela função
if result["erro"]:
    # Se houver um erro, exibe a mensagem de erro
    print(f"\nErro: {result['mensagem']}")
else:
    # Se a consulta for bem-sucedida, exibe as informações do endereço
    address_data = result["data"]
    print("\n--- Informações do Endereço (via BrasilAPI - Dados agregados, incluindo Correios) ---")
    print(f"CEP: {address_data.get('cep', 'N/A')}")
    print(f"Logradouro: {address_data.get('street', 'N/A')}")
    # A BrasilAPI v2 não fornece 'complemento' diretamente neste endpoint.
    # print(f"Complemento: {address_data.get('complemento', 'N/A')}")
    print(f"Bairro: {address_data.get('neighborhood', 'N/A')}")
    print(f"Localidade: {address_data.get('city', 'N/A')}")
    print(f"UF: {address_data.get('state', 'N/A')}")
    # Os campos abaixo (IBGE, GIA, DDD, SIAFI) não são fornecidos diretamente pela BrasilAPI v2 para consulta de CEP.
    # print(f"IBGE: {address_data.get('ibge', 'N/A')}")
    # print(f"GIA: {address_data.get('gia', 'N/A')}")
    # print(f"DDD: {address_data.get('ddd', 'N/A')}")
    # print(f"SIAFI: {address_data.get('siafi', 'N/A')}")
```
