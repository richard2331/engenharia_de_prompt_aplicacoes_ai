```python
# Initialize a list of numbers
numeros = [1, 2, 3]
# Initialize a variable to store the sum
soma = 0
# Initialize a counter for the loop
i = 0

# Loop through the list of numbers using a while loop
while i < len(numeros):
  # Add the current number to the sum
  soma += numeros[i]
  # Increment the counter
  i += 1
  # Print the list and the current value of 'i'
  print(numeros, end=" | i= ")
  print(i)
```
