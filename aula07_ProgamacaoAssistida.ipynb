mini projeto: Inversão de string


```python
# DEPOIS - Pythonico: slicing nativo da linguagem

def inverte_texto(texto):
    return texto[::-1]

palavras = ['Python', 'UDF', 'Inteligencia Artificial', 'Kadidja']
print('=== ABORDAGEM PYTHONICA (SLICING) ===')
for p in palavras:
    print(f"  '{p}' -> '{inverte_texto(p)}'")
```

mini projeto: Numero Primo



```python
# Logica Otimizada O(sqrt N)
import math

def eh_primo(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

numeros_teste = [2, 7, 10, 13, 100, 97]
print('=== LOGICA OTIMIZADA ===')
for num in numeros_teste:
    print(f'  {num} é primo? {eh_primo(num)}')
```

Projeto: Calculadora **Básica**

```python
# Define a função para somar dois números (a e b)
def somar(a, b):
    return a + b

# Define a função para subtrair dois números (a e b)
def subtrair(a, b):
    return a - b

# Define a função para multiplicar dois números (a e b)
def multiplicar(a, b):
    return a * b

# Define a função para dividir dois números (a e b)
def dividir(a, b):
    # Verifica se o divisor (b) é zero para evitar erro de divisão por zero
    if b == 0:
        # Lança um erro caso a divisão por zero seja tentada
        raise ValueError('Erro: divisao por zero nao e permitida!')
    return a / b

# Define a função principal da calculadora
def calculadora():
    # Dicionário que armazena as operações disponíveis, seus nomes e as funções correspondentes
    operacoes = {
        '1': ('Soma        (+)', somar),
        '2': ('Subtracao   (-)', subtrair),
        '3': ('Multiplicacao (x)', multiplicar),
        '4': ('Divisao     (/)', dividir),
    }

    # Imprime um cabeçalho para a calculadora
    print('=' * 40)
    print('       CALCULADORA PYTHONICA')
    print('=' * 40)

    # Inicia um loop infinito para manter a calculadora em execução até que o usuário saia
    while True:
        # Exibe as opções de operação para o usuário
        print('
Escolha a operacao:')
        for chave, (nome, _) in operacoes.items():
            print(f'  [{chave}] {nome}')
        print('  [0] Sair')

        # Solicita a escolha do usuário e remove espaços em branco extras
        escolha = input('
Opcao: ').strip()

        # Verifica se o usuário escolheu sair (opção '0')
        if escolha == '0':
            print('
Ate mais!')
            break # Sai do loop, encerrando a calculadora

        # Verifica se a opção escolhida é válida
        if escolha not in operacoes:
            print('Opcao invalida. Tente novamente.')
            continue # Volta para o início do loop para que o usuário escolha novamente

        # Tenta converter as entradas do usuário para números float
        try:
            a = float(input('Primeiro numero:  '))
            b = float(input('Segundo numero:   '))
        except ValueError:
            # Se a conversão falhar (entrada não numérica), imprime um erro
            print('Entrada invalida. Digite apenas numeros.')
            continue # Volta para o início do loop para que o usuário insira novamente

        # Pega o nome da operação e a função correspondente do dicionário
        nome_op, funcao = operacoes[escolha]

        # Tenta executar a operação
        try:
            resultado = funcao(a, b)
            # Verifica se o resultado é um número inteiro para formatar a saída adequadamente
            if resultado == int(resultado):
                print(f'
Resultado: {int(resultado)}')
            else:
                # Caso contrário, formata o resultado com 4 casas decimais
                print(f'
Resultado: {resultado:.4f}')
        except ValueError as e:
            # Captura e imprime erros específicos da operação (como divisão por zero)
            print(f'
{e}')

# Chama a função calculadora para iniciar o programa
calculadora()
```
