mini projeto: Inversão de string


```python
# DEPOIS - Pythonico: slicing nativo da linguagem

def inverte_texto(texto):
    return texto[::-1]

palavras = ['Python', 'UDF', 'Inteligencia Artificial', 'Kadidja']
print('=== ABORDAGEM PYTHONICA (SLICING) ===')
for p in palavras:
    print(f"  '{p}' -> '{inverte_texto(p)}'")
```

mini projeto: Numero Primo



```python
# Logica Otimizada O(sqrt N)
import math

def eh_primo(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

numeros_teste = [2, 7, 10, 13, 100, 97]
print('=== LOGICA OTIMIZADA ===')
for num in numeros_teste:
    print(f'  {num} é primo? {eh_primo(num)}')
```

```python
def somar(a, b):
    return a + b

def subtrair(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    if b == 0:
        raise ValueError('Erro: divisao por zero nao e permitida!')
    return a / b

def calculadora():
    operacoes = {
        '1': ('Soma        (+)', somar),
        '2': ('Subtracao   (-)', subtrair),
        '3': ('Multiplicacao (x)', multiplicar),
        '4': ('Divisao     (/)', dividir),
    }

    print('=' * 40)
    print('       CALCULADORA PYTHONICA')
    print('=' * 40)

    while True:
        print('\nEscolha a operacao:')
        for chave, (nome, _) in operacoes.items():
            print(f'  [{chave}] {nome}')
        print('  [0] Sair')

        escolha = input('\nOpcao: ').strip()

        if escolha == '0':
            print('\nAte mais!')
            break

        if escolha not in operacoes:
            print('Opcao invalida. Tente novamente.')
            continue

        try:
            a = float(input('Primeiro numero:  '))
            b = float(input('Segundo numero:   '))
        except ValueError:
            print('Entrada invalida. Digite apenas numeros.')
            continue

        nome_op, funcao = operacoes[escolha]

        try:
            resultado = funcao(a, b)
            if resultado == int(resultado):
                print(f'\nResultado: {int(resultado)}')
            else:
                print(f'\nResultado: {resultado:.4f}')
        except ValueError as e:
            print(f'\n{e}')

calculadora()
```
